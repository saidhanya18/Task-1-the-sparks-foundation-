# through all the graphs we conclude that as the number of study hours increases score also increases which indicates correct data .

# To prepare the data for our model 

```python
x=data.iloc[:, :-1].values
y=data.iloc[:, 1].values
```

# For testing and training the model 

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
# 20 percent of data is used to test 
```

# Training the algorithm

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

#from sklearn.ensemble import RandomForestRegressor
#regressor = RandomForestRegressor(n_estimators = 1000, random_state= 42 )

regressor.fit(x_train,y_train)
print("Training complete")
#80 percent od the dat is used to train 
```

# Prediction made by the model 

```python
print(y_test)
print("Prediction of Score")
y_pred = regressor.predict(x_test)
print(y_pred)
```

# Checking the accuracy of the model

```python
df=pd.DataFrame({'Actual': y_test,'Predicted': y_pred})
df
#dataframe is used to make a prediction as a csv file 
```

# Prediction with custom input 
what will be the score if the student studies for 9.25 hrs?

```python
hours = [[9.25]]
pred = regressor.predict(hours)
print(pred) 
# this predicts the score one can score when the person studies for 9.25 hours 
```

# Evaluating the model 

```python
from sklearn import metrics
print('Mean Absolute Error:',
     metrics.mean_absolute_error(y_test,y_pred))
#metrices helps us to find the mean absolute error 
```
